Label encoding to convert charecters to numeric

```python
cat_cols = ['airline', 'ch_code','from', 'stop', 'to', 'class','via']
for col in cat_cols:
    mapping = {}
    label = 0
    for category in data[col].unique():
        if category not in mapping:
            mapping[category] = label
            label += 1
    data[col] = data[col].apply(lambda x: mapping[x])
print(data)
```

```python
data=data.drop('ch_code',axis=1)
```

```python
train = data.sample(frac=0.8,random_state=357)
test = data.drop(train.index)
```

```python
X_train = train.drop('price', axis=1)
y_train = train['price']
X_test = test.drop(['price'], axis=1)
y_test = test['price']
```

```python
def distNeighbours(X_train,y_train,X_test,K):
    distance=[]
    for i in range(len(X_train)):
        eDistance=0
        for j in range(len(X_train.columns)):   
                eDistance+=round(np.sqrt(pow((X_train.iloc[i,j]-X_test[j]),2)),2)
        distance.append((eDistance,i,y_train.iloc[i]))
        distance=sorted(distance, key=lambda x: x[0])[0:K]
    return distance
```

```python
def predictOutput(X_train,y_train,X_test,K):
    neighbours=[]
    responses=[]
    for i in range(len(X_test)):
        neighbours.append(distNeighbours(X_train,y_train,X_test.iloc[i,:],K))
    for i in neighbours:
        mean=0
        for j in i:
            mean+=j[-1]
        mean=mean/K
        responses.append(mean)
    return responses
```

```python
output=predictOutput(X_train,y_train,X_test,3)
```
