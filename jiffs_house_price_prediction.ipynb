```python
train["land_size_sqm"] = train["land_size_sqm"].apply(lambda x: x.replace('sqm', '') if type(x) is str else x)
train["house_size_sqm"] = train["house_size_sqm"].apply(lambda x: x.replace('sqm', '') if type(x) is str else x)
train["distance_to_school_km"] = train["distance_to_school"].apply(lambda x: x.replace('km', '') if type(x) is str else x)
train["house_age_or_renovated"] = train["house_age"].apply(lambda x: x.replace('Years', '') if type(x) is str else x)
train["distance_to_supermarket_km"] = train["distance_to_supermarket"].apply(lambda x: x.replace('km', '') if type(x) is str else x)
```

```python
train.head()
```

```python
train=train.drop(['distance_to_school','house_age','distance_to_supermarket'],axis=1)
```

```python
cat_cols = ['large_living_room','parking_space','front_garden','swimming_pool','wall_fence','water_front','room_size']
for col in cat_cols:
    mapping = {}
    label = 0
    for category in train[col].unique():
        if category not in mapping:
            mapping[category] = label
            label += 1
    train[col] = train[col].apply(lambda x: mapping[x])
print(train)
```

```python
train_div = train.sample(frac=0.8,random_state=357)
test_div = train.drop(train.index)
```

```python
X_train = train_div.drop('property_value', axis=1)
y_train = train_div['property_value']
X_test = test_div.drop(['property_value'], axis=1)
y_test = test_div['property_value']
```

```python
def distNeighbours(X_train,y_train,X_test,K):
    distance=[]
    for i in range(len(X_train)):
        eDistance=0
        for j in range(len(X_train.columns)):   
                eDistance+=round(np.sqrt(pow((X_train.iloc[i,j]-X_test[j]),2)),2)
        distance.append((eDistance,i,y_train.iloc[i]))
        distance=sorted(distance, key=lambda x: x[0])[0:K]
    return distance
```

```python
def predictOutput(X_train,y_train,X_test,K):
    neighbours=[]
    responses=[]
    for i in range(len(X_test)):
        neighbours.append(distNeighbours(X_train,y_train,X_test.iloc[i,:],K))
    for i in neighbours:
        mean=0
        for j in i:
            mean+=j[-1]
        mean=mean/K
        responses.append(mean)
    return responses
```

```python
output=predictOutput(X_train,y_train,X_test,300)
```
